```python
#-*- coding:utf-8 -*-

# matplotlib 내 imread 를 사용하여 이미지 읽기

%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
import numpy as np

# matlplotlib은 이미지를 RGB 순으로 저장 
# opencv는 이미지를 BGR 순으로 저장

img = imread('q1.png')
plt.imshow(img)
plt.show()

# img2 = cv2.cvtColor(img,  cv2.COLOR_BGR2RGB)
# img[:,:,0] = RED
# img[:,:,1] = Green
# img[:,:,2] = BLUE


# 빨간색만 뽑기 

img_red = img.copy()
img_red[:,:,1] = 0 # 1: 초록색 없애기
img_red[:,:,2] = 0 # 2 파란색 없애기


# 초록색만 뽑기

img_green = img.copy() # 이렇게 해야 제대로 복사가 됩니다.
img_green[:,:,0] = 0 # 0: 빨간색 없애기
img_green[:,:,2] = 0 # 2: 파란색 없애기


# 파란색만 뽑기

img_blue  = img.copy() # 이렇게 해야 제대로 복사가 됩니다.
img_blue [:,:,0] = 0 # 0: 빨간색 없애기
img_blue [:,:,1] = 0 # 1: 초록색 없애기

r_max = np.max(img_red)
g_max = np.max(img_green)
b_max = np.max(img_blue)

print("R,G,B의 최대값: ", r_max, g_max, b_max)
print("\n")

r_mean = np.mean(img_red)
g_mean = np.mean(img_green)
b_mean = np.mean(img_blue)

print("R,G,B의 평균값: ", r_mean, g_mean, b_mean)

#  (R → G, G → B, B → R)

plt.subplot(1, 3, 1) # 1행 3열에서 1번째 열

# R -> G 
img2 = img.copy()
img2[:,:,0] = img2[:,:,1]
plt.imshow(img2)

plt.subplot(1, 3, 2) # 1행 3열에서 2번째 열

# G -> B 
img2 = img.copy()
img2[:,:,1] = img2[:,:,2]
plt.imshow(img2)

plt.subplot(1, 3, 3) # 1행 3열에서 3번째 열

# B -> R 
img2 = img.copy()
img2[:,:,2] = img2[:,:,0]
plt.imshow(img2)

plt.show()
```
