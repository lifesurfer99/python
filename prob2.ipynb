```python
#-*- coding:utf-8 -*-

# matplotlib  imread    

%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
import numpy as np

# matlplotlib  RGB   
# opencv  BGR  

img = imread('q1.png')
plt.imshow(img)
plt.show()

# img2 = cv2.cvtColor(img,  cv2.COLOR_BGR2RGB)
# img[:,:,0] = RED
# img[:,:,1] = Green
# img[:,:,2] = BLUE


#   

img_red = img.copy()
img_red[:,:,1] = 0 # 1:  
img_red[:,:,2] = 0 # 2  


#  

img_green = img.copy() #     .
img_green[:,:,0] = 0 # 0:  
img_green[:,:,2] = 0 # 2:  


#  

img_blue  = img.copy() #     .
img_blue [:,:,0] = 0 # 0:  
img_blue [:,:,1] = 0 # 1:  

r_max = np.max(img_red)
g_max = np.max(img_green)
b_max = np.max(img_blue)

print("R,G,B : ", r_max, g_max, b_max)
print("\n")

r_mean = np.mean(img_red)
g_mean = np.mean(img_green)
b_mean = np.mean(img_blue)

print("R,G,B : ", r_mean, g_mean, b_mean)

#  (R  G, G  B, B  R)

plt.subplot(1, 3, 1) # 1 3 1 

# R -> G 
img2 = img.copy()
img2[:,:,0] = img2[:,:,1]
plt.imshow(img2)

plt.subplot(1, 3, 2) # 1 3 2 

# G -> B 
img2 = img.copy()
img2[:,:,1] = img2[:,:,2]
plt.imshow(img2)

plt.subplot(1, 3, 3) # 1 3 3 

# B -> R 
img2 = img.copy()
img2[:,:,2] = img2[:,:,0]
plt.imshow(img2)

plt.show()
```
